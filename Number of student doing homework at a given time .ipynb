```python
class Solution:
    def busyStudent(self, startTime, endTime, queryTime):
        count = 0
        for start, end in zip(startTime, endTime):
            if start <= queryTime <= end:
                count += 1
        return count
```

```python
sol = Solution()

startTime = [1, 2, 3]
endTime = [3, 2, 7]
queryTime = 4
sol.busyStudent(startTime, endTime, queryTime)
```

```python
sol.busyStudent([4], [4], 4)
```
