```python
class Solution:
    def numTeams(self, rating):
        n = len(rating)
        count = 0
        
        # check all triplets
        for i in range(n):
            for j in range(i + 1, n):
                for k in range(j + 1, n):
                    if rating[i] < rating[j] < rating[k] or rating[i] > rating[j] > rating[k]:
                        count += 1
        return count
```

```python
sol = Solution()

rating = [2, 5, 3, 4, 1]
sol.numTeams(rating)
```

```python
sol.numTeams([2,1,3])
```

```python
sol.numTeams([1,2,3,4])
```
