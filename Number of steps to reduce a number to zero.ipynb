```python
class Solution:
    def numberOfSteps(self, num):
        steps = 0
        while num > 0:
            if num % 2 == 0:
                num //= 2
            else:
                num -= 1
            steps += 1
        return steps
```

```python
sol = Solution()

num = 14
sol.numberOfSteps(num)
```

```python
sol.numberOfSteps(8)    # Output: 4
sol.numberOfSteps(123)  # Output: 12
```
